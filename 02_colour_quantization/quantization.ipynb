```python
import cv2
import matplotlib.pyplot as plt
from image_quantization import quantize, show_images

%matplotlib inline
```

```python
img = cv2.imread("../lena.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image");
```

---

### Output with different levels of quantization

```python
# 16 colours

quant = quantize(img, levels = 16)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

show_images(rgb_img, quant)
```

```python
# 8 colours

quant = quantize(img, levels = 8)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

show_images(rgb_img, quant)
```

```python
# 4 colours

quant = quantize(img, levels = 4)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

show_images(rgb_img, quant)
```

```python
# 2 colours

quant = quantize(img, levels = 2)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

show_images(rgb_img, quant)
```
