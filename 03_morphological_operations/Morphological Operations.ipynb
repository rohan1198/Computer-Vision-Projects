```python
import cv2
import matplotlib.pyplot as plt
from morph import erosion, dilation, opening, closing, gradient

%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread("../lena.jpg")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure()
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[1].imshow(gray_img, cmap = plt.get_cmap("gray"));
```

---

```python
# Erosion

eroded_img = erosion(gray_img, iterations = 2)

plt.figure()
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
ax[1].imshow(eroded_img, cmap = plt.get_cmap("gray"));
```

---

```python
# Dilation

dilated_img = dilation(gray_img, iterations = 2)

plt.figure()
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
ax[1].imshow(dilated_img, cmap = plt.get_cmap("gray"));
```

---

```python
# Opening

opened_img = opening(gray_img, kernel_size = (5, 5))

plt.figure()
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
ax[1].imshow(opened_img, cmap = plt.get_cmap("gray"));
```

---

```python
# Closing

closed_img = closing(gray_img, kernel_size = (5, 5))

plt.figure()
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
ax[1].imshow(closed_img, cmap = plt.get_cmap("gray"));
```

---

```python
# Gradient

gradient_img = gradient(gray_img, kernel_size = (3, 3))

plt.figure()
fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
ax[1].imshow(gradient_img, cmap = plt.get_cmap("gray"));
```
