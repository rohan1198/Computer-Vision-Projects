```python
import cv2
import matplotlib.pyplot as plt
from equalize import simple_equalization, adaptive_equalization, show

%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread("../lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image");
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap = plt.get_cmap("gray"))
plt.title("Grayscale Image");
```

---

```python
# Simple Histogram Equalization

sim_eq_img = simple_equalization(gray_img)
show(img, gray_img, sim_eq_img)
```

---

```python
# Adaptive Histogram Equalization

ada_eq_img = adaptive_equalization(gray_img,  clip = 1.1, grid = 8)
show(img, gray_img, ada_eq_img)
```
