```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread("../Graphics/Chapter01/Figure1.03.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
```

```python
h, s, v = cv2.split(hsv_img)

fig, ax = plt.subplots(1, 3)

ax[0].imshow(h)
ax[1].imshow(s)
ax[2].imshow(v)
plt.show()
```

```python
print(h.shape)
print(s.shape)
print(v.shape)
```

```python
neg_v = 255 - v
```

```python
value_img = cv2.merge((neg_v, neg_v, neg_v))
plt.imshow(v)
```

```python
# Horizontal Flip

flipped_img = value_img[:, ::-1, :]
plt.imshow(flipped_img)
```

```python
canvas = np.zeros((img.shape[0], img.shape[1]*2, img.shape[2]), dtype = np.uint8)

print(canvas.shape)
plt.imshow(canvas)
```

```python
canvas[:, :img.shape[1], :] = img

plt.imshow(canvas)
```

```python
canvas[:, flipped_img.shape[1]:, :] = flipped_img

plt.imshow(canvas)
```
