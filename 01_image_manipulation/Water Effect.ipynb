```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
img.shape
```

```python
canvas = np.zeros((img.shape[0] * 2, img.shape[1], 3), dtype = np.uint8)
canvas.shape
```

```python
plt.imshow(canvas)
```

```python
canvas[:img.shape[0], :, :] = img
plt.imshow(canvas)
```

```python
# Vertical flip

flipped_img = img[::-1, :, :]
plt.imshow(flipped_img)
```

```python
canvas[img.shape[0]:, :, :] = flipped_img
```

```python
plt.imshow(canvas)
```
