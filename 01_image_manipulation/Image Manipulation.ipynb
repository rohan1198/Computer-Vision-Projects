```python
import cv2
import matplotlib.pyplot as plt
from manipulation import scale, rotate, translate, crop, show_images

%matplotlib inline
```

```python
# Reading the image

img = cv2.imread("../lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

---

```python
# Scaling

scaled_img = scale(img, (250, 250))
show_images(img, scaled_img)
```

```python
print("Original Image shape: ", img.shape)
print("Scaled Image shape", scaled_img.shape)
```

---

```python
# Rotate

rotated_img = rotate(img, 45)
show_images(img, rotated_img)
```

---

```python
# Translate

translated_img = translate(img, [100, 200])
show_images(img, translated_img)
```

---

```python
# Crop

cropped_img = crop(img, [100, 400], [100, 400])
show_images(img, cropped_img)
```
