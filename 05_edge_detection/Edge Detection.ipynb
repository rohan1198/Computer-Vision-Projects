```python
import cv2
import matplotlib.pyplot as plt
from detect import edge_detection, auto_edge_detection, show

%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread("../lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image");
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap = plt.get_cmap("gray"))
plt.title("Grayscale Image");
```

---

```python
# Simple Edge Detection

simple_edge_img = edge_detection(gray_img, low_thresh = 30, high_thresh = 150)
show(img, simple_edge_img)
```

---

```python
# Automatic Edge Detection

auto_edge_img = auto_edge_detection(gray_img, sigma = 0.33)
show(img, auto_edge_img)
```
