```python
import cv2
import matplotlib.pyplot as plt
from blur import avg_blur, gaus_blur, med_blur, bil_blur, show

%matplotlib inline
```

```python
img = cv2.imread("../lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image");
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap = plt.get_cmap("gray"))
plt.title("Grayscale Image");
```

---

```python
# Average Blurring

avg_blurred_img = avg_blur(gray_img, kernel_size = 7)
show(img, avg_blurred_img)
```

---

```python
# Gaussian Blurring

gaus_blurred_img = gaus_blur(gray_img, kernel_size = 7)
show(img, gaus_blurred_img)
```

---

```python
# Median Blurring

med_blurred_img = med_blur(gray_img, kernel_size = 7)
show(img, avg_blurred_img, method = "median")
```

---

```python
# Bilateral Blurring

bil_blurred_img = bil_blur(gray_img, diameter = 11, colour = 61, space = 39)
show(img, avg_blurred_img)
```
